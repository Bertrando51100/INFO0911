```python
# Comparaison des bibliothèques Pillow et OpenCV
from PIL import Image
import numpy as np
import cv2
import matplotlib.pyplot as plt

def charger_image_pillow(chemin):
    """
    Charge une image depuis le chemin spécifié en utilisant Pillow.
    Retourne la matrice de pixels sous forme de tableau NumPy.
    """
    img = Image.open(chemin)
    return np.array(img)

def charger_image_cv2(chemin, convertir_rgb=True):
    """
    Charge une image depuis le chemin spécifié en utilisant OpenCV.
    Retourne la matrice de pixels sous forme de tableau NumPy.
    """
    img = cv2.imread(chemin)  # Charge l'image en BGR
    if convertir_rgb:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convertit en RGB pour correspondre à Pillow
    return img

# Exemple d'utilisation
if __name__ == "__main__":
    chemin_image = "BD_images_resized/Montres_37/image371.jpg"  # Remplacez par le chemin de votre image

    # Charger les images avec Pillow et OpenCV
    image_pillow = charger_image_pillow(chemin_image)
    image_cv2_1 = charger_image_cv2(chemin_image)
    image_cv2_2 = charger_image_cv2(chemin_image, convertir_rgb=False)

    # Afficher les deux images côte à côte
    plt.figure(figsize=(15, 5))

    plt.subplot(1, 3, 1)
    plt.title("Image chargée avec Pillow")
    plt.imshow(image_pillow)
    plt.axis("off")

    plt.subplot(1, 3, 2)
    plt.title("Image chargée avec OpenCV convertie en RGB")
    plt.imshow(image_cv2_1)
    plt.axis("off")
    
    plt.subplot(1, 3, 3)
    plt.title("Image chargée avec OpenCV non convertie")
    plt.imshow(image_cv2_2)
    plt.axis("off")

    plt.show()
```

```python
# Inspection des propriétés d'une image

def informations_image(image):
    """
    Affiche les informations sur une image :
    - Dimensions (hauteur, largeur, nombre de canaux)
    - Type de données (dtype)
    - Valeurs minimale et maximale des pixels
    """
    # Dimensions de l'image
    dimensions = image.shape
    hauteur, largeur = dimensions[:2]
    canaux = dimensions[2] if len(dimensions) == 3 else 1  # Vérifie si l'image est en niveaux de gris ou en couleur

    # Type de données
    type_donnees = image.dtype

    # Valeurs minimale et maximale des pixels
    valeur_min = image.min()
    valeur_max = image.max()

    # Affichage des informations
    print(f"Dimensions : {hauteur}x{largeur}, Nombre de canaux : {canaux}")
    print(f"Type de données : {type_donnees}")
    print(f"Valeur minimale des pixels : {valeur_min}")
    print(f"Valeur maximale des pixels : {valeur_max}")

## informations_image
informations_image(image_pillow)
```
